```python
x_sq = x**2
x_cube = x**3
x_sqrt = np.sqrt(x)
x_exp  = np.exp(x)
```

```python
plt.plot(x,x_sq, color='green', marker='o',linestyle='dashed',
     linewidth=2, markersize=2)


plt.plot(x,x_cube,'b--')
plt.plot(x,x_exp,'r.')
plt.plot(x,x,'co',linewidth=2,markersize=2)
```
