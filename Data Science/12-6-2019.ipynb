```python
fig = plt.figure(figsize=(10,4))

#ax = fig.add_axes([0,0,1,1])
ax1 = fig.add_axes([0.1,0.1,0.4,0.3])
ax2 = fig.add_axes([0.1,0.6,0.4,0.3])
ax3 = fig.add_axes([0.6,0.6,0.4,0.3])
ax4 = fig.add_axes([0.6,0.1,0.4,0.3])


ax1.plot(x,6*x*2+3*x+5,c='r',lw='2',label='axes1')
ax1.legend()

ax2.plot(x,x,'g--',lw='2',label='axes2')
ax2.plot(x,x**2,'m.',lw='2',label='axes2')
ax2.plot(x,x**3,'b.',lw='2',label='axes2')



plt.show()
```

```python
fig = plt.figure(figsize=(5,3))
ax = fig.add_axes([0,0,1,1])
ax1 = fig.add_axes([0.3,0.3,0.5,0.5])
y=x
ax1.plot(x,x**2,'m-',lw='2',label='axes1')
ax1.plot(x,x**3,'g.',lw='2',label='axes1')

ax.plot(y,x,'r-',lw='2',label='axes')
plt.show()
```
